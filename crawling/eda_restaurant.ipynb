```python
import pandas as pd
import re
```

```python
with open("restaurant.csv", "r", encoding="CP949", errors="replace") as f:
    data = f.read()

with open("restaurant_utf8.csv", "w", encoding="UTF-8") as f:
    f.write(data)

df = pd.read_csv("restaurant_utf8.csv", encoding="UTF-8")
```

```python
df.columns
```

```python
df_selected = df[["상세영업상태명","전화번호", 
                "소재지면적","소재지우편번호","지번주소","도로명주소", 
                "도로명우편번호", "사업장명","업태구분명", "좌표정보(X)", "좌표정보(Y)"]]
```

```python
df_seo = df_selected[df_selected["지번주소"].str.contains("서교동", na=False)]
```

```python
df_seo.shape[0]
```

```python
df_seo.loc[df_seo["상세영업상태명"]=="영업"].shape[0]
```

```python
df_seo_fil = df_seo.loc[df_seo["상세영업상태명"]=="영업"]
```

```python
df_hap = df_selected[df_selected["지번주소"].str.contains("합정", na=False)]
```

```python
df_hap.loc[df_hap["상세영업상태명"]=="영업"].shape[0]
```

```python
df_hap_fil = df_hap.loc[df_hap["상세영업상태명"]=="영업"]
```

```python
df_combined = pd.concat([df_seo_fil, df_hap_fil], ignore_index=True)
```

```python
df_combined.head()
```

```python
df_combined["업태구분명"].unique()
```

```python
df_combined[df_combined["업태구분명"]=="라이브카페"] # 라이브 카페는 술집의 느낌이여서 제외하지 않음
```

```python
df_combined[df_combined["업태구분명"]=="기타"] # 기타에 카페 같은것이 섞여 있고 분류하기가 어려워서 제외함 / 제외하여도 크게 문제가 없을 듯
```

```python
df_combined_fil = df_combined[~df_combined["업태구분명"].isin(["까페", "출장조리","기타"])]
```

```python
df_combined_fil
```

```python
df_final = df_combined_fil[["소재지면적","지번주소","도로명주소" ,"사업장명","업태구분명","좌표정보(X)","좌표정보(Y)"]]
```

```python
df_final
```

```python
df_final["사업장명"] = df_final["사업장명"].str.replace(r"\(.*?\)", "", regex=True).str.strip() # 괄호내 영문 혹은 문제가 되는 텍스트가 존재하여 제거를 함.
```

```python
df_final.to_csv("restaurant_df.csv", index = False)
```
