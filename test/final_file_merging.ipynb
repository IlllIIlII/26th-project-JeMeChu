```python
import pandas as pd

menu = pd.read_csv("restaurant_final.csv", encoding="utf-8-sig")
final = pd.read_csv("preprocessed_naver.csv", encoding="utf-8-sig")
```

```python
final.head(1)
```

```python

final = final.merge(menu[['사업장명', '메뉴']], left_on="사업장명", right_on="사업장명", how='left')
```

```python
final.head(1)
```

```python
final.rename(columns={
    "사업장명": "name",
    "지번주소": "jibun_address",
    "도로명주소": "road_address",
    "업태구분명": "category",
    "전화번호": "phone",
    "운영시간": "business_hours",
    "총 리뷰 개수": "review_count",
    "소개": "description",
    "주차 정보": "parking",
    "좌석 정보": "seat_info",
    "최신 300개 리뷰": "latest_reviews",
    "소재지면적": "size",
    "편의시설 및 서비스" : "facilities",
    "이런점이 좋았어요": "very_good",
    "tfidf_features": "tfidf_features",
    "좌표정보(X)": "latitude",
    "좌표정보(Y)": "longitude",
    "메뉴" : "menu"
}, inplace=True)
```

```python
# 📌 변경된 데이터 저장
output_path = "final_updated.csv"
final.to_csv(output_path, index=False, encoding="utf-8-sig")
```
