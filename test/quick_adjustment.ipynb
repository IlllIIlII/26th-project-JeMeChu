```python
import pandas as pd

df = pd.read_csv('restaurant_temp.csv')
```

```python
# 조건 적용: "Processed"가 "Yes"이고 "메뉴"가 빈 리스트인 경우
mask = (df["Processed"] == "Yes") & (df["메뉴"] == "[]")

# "Processed"와 "메뉴" 값을 빈 문자열로 변경
df.loc[mask, ["Processed", "메뉴"]] = ""
```
