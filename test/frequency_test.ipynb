```python
from kiwipiepy import Kiwi
from kiwipiepy.utils import Stopwords
import re

kiwi = Kiwi()
stopwords = Stopwords()

# 전처리 코드

def do_Kr_preprocessing(text):
    filtered_content = re.sub('[^\s\w\d]',' ', text)
    kiwi_tokens = kiwi.tokenize(filtered_content, stopwords=stopwords)
    Noun_words = []
    for token in kiwi_tokens:
        if 'NN' in token.tag:
            Noun_words.append(token.form)
    # 길이가 1인 단어 제거 하기
    final_Noun_words = []
    for word in Noun_words:
        if len(word)>1:
            final_Noun_words.append(word)
    return final_Noun_words
```

```python
test_df = df.loc[:10,["사업장명","최신 300개 리뷰"]]
```

```python
test_df
```

```python
test_df["processed_reviews"] = test_df["최신 300개 리뷰"].apply(do_Kr_preprocessing)
```

```python
from collections import Counter

# Apply frequency analysis to each row
test_df["top_10_words"] = test_df["processed_reviews"].apply(lambda x: Counter(x).most_common(10))
```

```python
from kiwipiepy import Kiwi
from kiwipiepy.utils import Stopwords
import re
from collections import Counter


kiwi = Kiwi()
stopwords = Stopwords()

# 전처리 코드

def do_Kr_verb_preprocessing(text):
    filtered_content = re.sub('[^\s\w\d]', ' ', text)  # Remove special characters
    kiwi_tokens = kiwi.tokenize(filtered_content, stopwords=stopwords)
    Verb_words = []

    for token in kiwi_tokens:
        if token.tag == 'VA':  # Extract both verbs and adjectives
            Verb_words.append(token.form)

    # Remove single-character words
    final_Verb_words = [word for word in Verb_words if len(word) > 1]

    return final_Verb_words


test_df["processed_reviews"] = test_df["최신 300개 리뷰"].apply(do_Kr_verb_preprocessing)

# Apply frequency analysis to each row
test_df["top_10_words"] = test_df["processed_reviews"].apply(lambda x: Counter(x).most_common(10))
```
